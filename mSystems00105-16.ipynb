### Parse the nitrate, nitrite and nitric KOs

We are going to search for all KOs that have nitrate, nitrite and nitric + reductase in their KO name. We will use those string later with PICRUSt.

````python
file = [l for l in open('ko_13_5_precalculated.tab')]
head = file[0].split('\t')
for nname in ['nitrite', 'nitric', 'nitrate']:
   nitrate_idx = {head[i]: n for i, n in enumerate(file[-2].split('\t')) if nname in n.lower() and 'reductase' in n.lower()}
   print nname, ','.join(nitrate_idx)
````

## Commands for Oligotyping.

Note that these need the actual sequence file, which can be found [here](ftp://ftp.microbio.me/AmericanGut/rounds-1-25/01-raw/). Also, that you need to check the entropy plots to select the `-c` and other analytical parameters. Keep in mind that those selections are based on personal evaluation.

Below the commands for *Pseudomonas*, which are the same for Streptococcus but just with a different `-c`.

```bash
sed -e 's/^\(>[^[:space:]]*\).*/\1/' sequences.fna > AG_renamed.fasta

read_fasta -i AG_renamed.fasta | grab -i -E oral_sample_names.txt -k SEQ_NAME | write_fasta -o oral_reads.fasta -x

vsearch -usearch_global oral_reads.fasta -db Pseudomonas_otus.fasta -strand plus -id 0.97 -uc oral_Pseudomonas.uc

cut -f 9 oral_Pseudomonas.uc > pseudomonas_reads_ids.txt

seqtk subseq oral_reads.fasta pseudomonas_reads_ids.txt > pseudomonas_reads1.fasta

read_fasta -i pseudomonas_reads1.fasta | plot_distribution -k SEQ_LEN -t png -o plot.png -x

read_fasta -i pseudomonas_reads1.fasta | grab -e 'SEQ_LEN>=125' | grab -e 'SEQ_LEN<=10000000' | write_fasta -o pseudomonas_reads2.fasta -x

fastx_trimmer -i pseudomonas_reads2.fasta -o pseudomonas_reads3.fasta -f 1 -l 125

o-get-sample-info-from-fasta pseudomonas_reads3.fasta

entropy-analysis pseudomonas_reads3.fasta

oligotype pseudomonas_reads3.fasta pseudomonas_reads3.fasta-ENTROPY -c 7 -a 3 -s 80
```
